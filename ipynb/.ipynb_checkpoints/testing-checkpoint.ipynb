```python
def findGroup( self, answer, group, codes):
    # go through answer one by one, see if the exact group exists, stop when found
    for i in range(0, len(answer) - len(codes) -1):
        if (answer.ix[i:i+len(codes)-1, "diag"] == codes).all():
            #winner
            answer.ix[i:i+len(codes)-1, "Grouping"] = group
            return answer
```
